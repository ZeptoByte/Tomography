```python
import numpy as np
import matplotlib.pyplot as plt
N_Injected = int(1e8)

kinetic_energy = 10**(np.random.uniform(0,5,N_Injected))
cos_zenith = np.random.uniform(0,1,N_Injected)
```

```python
x_space = np.geomspace(.01,1e5,50)
y_space = .2*np.exp(-2.7*(x_space))
weights = np.exp(-2.7*np.log10(kinetic_energy)+np.log(kinetic_energy))
plt.figure()
#plt.hist(kinetic_energy, weights=weights, bins=np.geomspace(.01,1e5,50), histtype='step', density=True)
plt.plot(x_space, y_space)
plt.xscale('log')
plt.yscale('log')
plt.ylim(bottom=1e-20)
plt.show()
```

```python
x_space = np.linspace(0,1,100)
y_space = 3*x_space**2
weights = cos_zenith**2
plt.figure()
plt.hist(cos_zenith, weights=weights,bins=np.linspace(0,1,50), histtype='step', density=True)
plt.plot(x_space, y_space)
#plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the size of the sample
sample_size = 1000

# Define the exponent of the power-law distribution
exponent = -2.7

# Generate samples from a power-law distribution
samples = np.random.power(exponent, sample_size)

# Plot the histogram of the samples
plt.hist(samples, bins=50, density=True, alpha=0.7, color='blue')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title(f'Power Law Distribution (Exponent = {exponent})')

# Show the plot
plt.show()
```
